````python
from openai import OpenAI

# OpenAIクライアントのインスタンスを作成
client = OpenAI(api_key='')  # APIキーを設定
````

````python
# シンプルなテキスト分割関数
def split_text(text, chunk_size, chunk_overlap):
    chunks = []
    start = 0
    text_length = len(text)
    while start < text_length:
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start += (chunk_size - chunk_overlap)
    return chunks

# サンプルテキストを分割（sample_text を利用）
chunks = split_text(text, chunk_size=400, chunk_overlap=100)
````

````python
# 技術知識抽出用のプロンプトテンプレート（技術的手法、プロセス、パラメータ、用語、関係性をJSON形式で抽出）
prompt_template = (
    "以下のテキストから、技術的な手法、プロセス、パラメータ、用語、及びそれらの関係性を抽出し、JSON形式で返してください。なお用語については頻出度の高い上位5つの用語を抽出してください。テキスト:\n\n{chunk}"
)
````

````python
# 任意のチャンクのみを使用
chunk = chunks[2]
prompt = prompt_template.format(chunk=chunk)

response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "あなたは技術知識抽出の専門家です。"},
            {"role": "user", "content": prompt}
        ],
        max_tokens=1000,
        temperature=0.2
    )

````

````python

# 各チャンクに対してOpenAI APIを呼び出し、結果を表示する
responses = []
for i, chunk in enumerate(chunks):
    prompt = prompt_template.format(chunk=chunk)
    print(f"Processing chunk {i+1}:\n{chunk}\n")
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "あなたは技術知識抽出の専門家です。"},
                {"role": "user", "content": prompt}
            ],
            max_tokens=1000,
            temperature=0.2
        )
        extracted_json = response.choices[0].message.content.strip()
        print("Response:")
        print(extracted_json)
        responses.append(extracted_json)
    except Exception as e:
        print(f"Error processing chunk {i+1}: {e}")

print("\n=== 全チャンクの応答 ===")
for i, resp in enumerate(responses):
    print(f"Chunk {i+1} response:")
    print(resp)
    print("-" * 40)
````

````python
import json

# 各応答からマークダウン部分（```json ... ```）を取り除き、JSONをパースする関数
def parse_response(resp_str):
    # 改行で分割
    lines = resp_str.splitlines()
    # 最初の行が ```json なら除去
    if lines and lines[0].startswith("```"):
        lines = lines[1:]
    # 最後の行が ``` なら除去
    if lines and lines[-1].startswith("```"):
        lines = lines[:-1]
    json_str = "\n".join(lines)
    return json.loads(json_str)

parsed_responses = [parse_response(r) for r in responses]

# 全応答を1つのJSON配列として表示
all_responses_json = json.dumps(parsed_responses, ensure_ascii=False, indent=2)
print(all_responses_json)
````

````python
import json
import os
from pyvis.network import Network

# ナレッジDBを保存するファイル名
KNOWLEDGE_DB_FILE = "knowledge_db.json"

# 既存のナレッジDBを読み込む（存在しない場合は初期化する）
def load_knowledge_db():
    if os.path.exists(KNOWLEDGE_DB_FILE):
        with open(KNOWLEDGE_DB_FILE, "r", encoding="utf-8") as f:
            return json.load(f)
    else:
        # 初期状態として、ドキュメントとグラフ情報を保持する構造を作成
        return {"documents": [], "graph": {"nodes": [], "edges": []}}

# ナレッジDBを保存する
def save_knowledge_db(db):
    with open(KNOWLEDGE_DB_FILE, "w", encoding="utf-8") as f:
        json.dump(db, f, ensure_ascii=False, indent=2)
````

````python
# 抽出した技術ナレッジ情報および全文からナレッジDBを更新する関数
def update_knowledge_db(db, extracted_knowledge, full_text, doc_id):
    # ドキュメントごとのエントリを作成
    doc_entry = {
        "id": doc_id,
        "full_text": full_text,
        "extracted_knowledge": extracted_knowledge
    }
    db["documents"].append(doc_entry)
    
    # ドキュメント自体を表すノードを追加（まだなければ）
    if not any(node["id"] == doc_id for node in db["graph"]["nodes"]):
        db["graph"]["nodes"].append({
            "id": doc_id,
            "label": doc_id,
            "group": "document"
        })
    
    # 各カテゴリごとにノードとエッジを追加
    # ここでは「技術的な手法」「プロセス」「パラメータ」「用語」を扱います
    categories = {
        "技術的な手法": "technique",
        "プロセス": "process",
        "パラメータ": "parameter",
        "用語": "term"
    }
    for key, group in categories.items():
        items = extracted_knowledge.get(key, [])
        # items が dict であればキー一覧に変換
        if isinstance(items, dict):
            items = list(items.keys())
        for item in items:
            # ノードが存在しなければ追加
            if not any(node["id"] == item for node in db["graph"]["nodes"]):
                db["graph"]["nodes"].append({
                    "id": item,
                    "label": item,
                    "group": group
                })
            # ドキュメントから各項目へのエッジを追加
            db["graph"]["edges"].append({
                "source": doc_id,
                "target": item,
                "label": key
            })
    
    # 「関係性」情報を処理
    # ここでは、関係性は dict として、各ソースキーから対象リストを持つものとする
    relationships = extracted_knowledge.get("関係性", {})
    if isinstance(relationships, dict):
        for src, targets in relationships.items():
            # ソースノードがなければ追加（グループは "relation" としておく）
            if not any(node["id"] == src for node in db["graph"]["nodes"]):
                db["graph"]["nodes"].append({
                    "id": src,
                    "label": src,
                    "group": "relation"
                })
            for target in targets:
                if not any(node["id"] == target for node in db["graph"]["nodes"]):
                    db["graph"]["nodes"].append({
                        "id": target,
                        "label": target,
                        "group": "relation"
                    })
                # 関係性のエッジ（ラベルは「関連」など）
                db["graph"]["edges"].append({
                    "source": src,
                    "target": target,
                    "label": "関連"
                })
    return db

````

````python
def visualize_graph(db, output_filename="graph.html"):
    # notebook=False を指定して、Jupyter Notebook 用の処理を無効にします
    net = Network(height="600px", width="100%", directed=False, notebook=False)
    
    # DBのノードを追加
    for node in db["graph"]["nodes"]:
        net.add_node(node["id"], label=node["label"], group=node["group"])
    
    # DBのエッジを追加
    for edge in db["graph"]["edges"]:
        net.add_edge(edge["source"], edge["target"], label=edge["label"])
    
    # HTML を生成してファイルに保存
    html = net.generate_html()
    with open(output_filename, "w", encoding="utf-8") as f:
        f.write(html)
    print(f"グラフが '{output_filename}' として保存されました。ブラウザでご確認ください。")
````

````python
# ここからは実際の実行部分（メイン処理）
if __name__ == "__main__":
    # 例として、前回の技術知識抽出の結果（ダミー）を用意します。
    dummy_extracted_knowledge = {
        "技術的な手法": ["逆浸透膜法", "化学蒸着法"],
        "プロセス": ["送液", "昇圧", "膜形成"],
        "パラメータ": ["透過水量", "耐薬品性", "膜厚"],
        "用語": ["逆浸透膜", "酸化シリコン膜"],
        "関係性": {
            "逆浸透膜法": ["逆浸透膜", "昇圧"],
            "化学蒸着法": ["酸化シリコン膜", "膜形成"]
        }
    }
    # 抽出元の全文（ここでは先ほどのサンプルテキストを利用）
    full_text = sample_text  # sample_text は前回定義したサンプルテキスト
    # ドキュメントIDを設定（例： "doc_1"）
    doc_id = "doc_1"
    
    # ナレッジDBの読み込み
    db = load_knowledge_db()
    
    # ナレッジDBの更新（新たなドキュメントエントリを追加）
    db = update_knowledge_db(db, dummy_extracted_knowledge, full_text, doc_id)
    
    # 更新したDBを保存
    save_knowledge_db(db)
    
    # グラフの可視化（HTMLファイル "graph.html" として出力）
    visualize_graph(db, output_filename="graph.html")
    print("グラフが 'graph.html' として出力されました。ブラウザでご確認ください。")
````
